```python
def text_to_sql_query(text):
    model = genai.GenerativeModel('gemini-1.5-flash')
    schema = """
    Schema:
    Table: invoices
    - invoice_id: INT
    - user_id: INT
    - invoice_date: DATE
    - amount: DECIMAL
    """
    prompt = f"{schema}\n\nConvert the following to an SQL query:\n{text}\n\nSQL:"
    response = model.generate_content(prompt)
    return response.text.strip()

# Example usage:
text_input = "what is the amount of my last invoice"
sql_query = text_to_sql_query(text_input)
sql_query
```
